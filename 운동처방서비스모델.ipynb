```python
train = result1[['BMI','Weight','age','bending_forward','body_fat','diastolic_bp','grip_power','height','measure_time','sex','sit_up','step_inspection','step_recover_time','systolic_bp']]
label = r
```

```python
from sklearn.preprocessing import StandardScaler  
from sklearn.model_selection import train_test_split 
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression

m = list()
for i in range(268):
    smote = SMOTE(random_state=0)
    train_over,label_over = smote.fit_sample(train,label[r.columns[i]])
    print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', train.shape, label[r.columns[i]].shape)
    print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', train_over.shape, label_over.shape)
    print('SMOTE 적용 후 레이블 값 분포: \n', label_over.value_counts())
    train_features, test_features, train_labels, test_labels = train_test_split(train_over, label_over, stratify=label_over)
    print(len(train_features), len(test_features))
    scaler = StandardScaler()  
    train_features = scaler.fit_transform(train_features)  
    test_features = scaler.transform(test_features)
    model = LogisticRegression()
    model.fit(train_features, train_labels)
    print("train_score: ",model.score(train_features, train_labels))
    print("test_score: ",model.score(test_features, test_labels))
    m.append(model)
```
